## Predicting survivability of passengers

In this project I will try to accurately predict if passengers survived the sinking of the Titanic or not. Since the goal of this project is to predict a label (survived or not) then it means I am dealing with a binary classification problem.

### Dataset

For this Kaggle challenge we have a dataset split into two groups: training set and test set. These sets contain Titanic passengers' information but only the training set indicates for each passenger if he/she survived the disaster.

The variables in this dataset are the following:

| Variable | Definition | Key |
| --- | --- | ---|
| survival | Survival | 0 = No, 1 = Yes |
| pclass | Ticket class | 1 = 1st, 2 = 2nd, 3 = 3rd
| sex | Sex |
| Age | Age in years | 
| sibsp | # of siblings / spouses aboard the Titanic |
| parch | # of parents / children aboard the Titanic |	
| ticket | Ticket number |	
| fare | Passenger fare |	
| cabin | Cabin number |
| embarked | Port of Embarkation | C = Cherbourg, Q = Queenstown, S = Southampton

**Notes about the variables:**

*pclass:* A proxy for socio-economic status (SES)

- 1st = Upper

- 2nd = Middle

- 3rd = Lower

*age:* Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5

*sibsp:* The dataset defines family relations in this way:
- Sibling = brother, sister, stepbrother, stepsister
- Spouse = husband, wife (mistresses and fianc√©s were ignored)

*parch:* The dataset defines family relations in this way:
- Parent = mother, father
- Child = daughter, son, stepdaughter, stepson

Some children travelled only with a nanny, therefore parch=0 for them.

### Evaluation metric

The evaluation metric set by [Kaggle](https://www.kaggle.com/c/titanic/overview/evaluation) in this competition is **accuracy**.

#### 1.1.4 Cabin

With so many missing values, my first instinct was to delete the `Cabin` variable but then I searched on Google and it was pointed out that the first letter of a cabin number refers to the deck that the cabin belonged to. Therefore I will extract these letters and create a new feature that indicates the passenger's deck. For the missing values, I will assign the letter $M$ that indicates that it is a missing value. Finally, I will remove the `Cabin` variable since it has been replaced with this new feature.

```python
def getFirstLetter(cabin):
    # If cabin number is not NaN
    if pd.notnull(cabin):
        letter = cabin[0]
    else:
        letter = 'M' # M stands for missing
        
    return letter

# Create 'Deck' variable by extracting the first letter in the 'Cabin' variable
df_train['Deck'] = df_train['Cabin'].apply(getFirstLetter)
df_test['Deck'] = df_test['Cabin'].apply(getFirstLetter)

# Remove Cabin from both sets
df_train.drop(['Cabin'], axis=1, inplace=True)
df_test.drop(['Cabin'], axis=1, inplace=True)
```

### 2. Feature engineering

#### 2.1 Creating `Title` feature from `Name` variable

By looking at the variable `Name` we see that the passengers' names contain titles like Mr., Mrs., Master., etc. These titles give us some useful information about sex(Mr = male, Mrs = married female), age (Miss is usually younger than Mrs) and profession (Master indicates profession and hence social status) which in the end could tell us something more about the survival chance for each passenger. 

I will extract these titles from passengers' names and check what titles we have in the dataset.

```python
def getTitleFromName(name):
    title = re.search('[a-zA-z]+(?=\.)', name).group()
    
    return title

# Get title from name and create new feature in training and test sets
df_train['Title'] = df_train['Name'].apply(getTitleFromName)
df_test['Title'] = df_test['Name'].apply(getTitleFromName)

# Check how many titles there are
df_train['Title'].value_counts()
```

```python
df_test['Title'].value_counts()
```

As we can see there are many titles but most of them appear very rarely, so I will put these rare titles into groups. Titles like Dr, Rev, Col, Major, Capt will be put into the 'Officer' group. Titles such as Dona, Jonkheer, Countess, Sir, Lady, Don were usually entitled to the aristocrats and so these titles will be put into the 'Aristocrat' group. I will also replace Mlle and Ms with Miss and Mme by Mrs as these are French titles.

I will remove the `Name` variable since it will be replaced with `Title`.

```python
# Replace Dr, Rev, Col, Major, Capt with 'Officer' 
df_train['Title'].replace(to_replace=['Dr', 'Rev', 'Col', 'Major', 'Capt'], value='Officer', inplace=True)
df_test['Title'].replace(to_replace=['Dr', 'Rev', 'Col', 'Major', 'Capt'], value='Officer', inplace=True)

# Replace Dona, Jonkheer, Countess, Sir, Lady, Don with 'Aristocrat'
df_train['Title'].replace(to_replace=['Dona', 'Jonkheer', 'Countess', 'Sir', 'Lady', 'Don'], value='Aristocrat', inplace=True)
df_test['Title'].replace(to_replace=['Dona', 'Jonkheer', 'Countess', 'Sir', 'Lady', 'Don'], value='Aristocrat', inplace=True)

# Replace Mlle and Ms with Miss and Mme with Mrs
df_train['Title'].replace({'Mlle': 'Miss', 'Ms': 'Miss', 'Mme': 'Mrs'}, inplace=True)
df_test['Title'].replace({'Mlle': 'Miss', 'Ms': 'Miss', 'Mme': 'Mrs'}, inplace=True)

# Remove 'Name' variable
df_train = df_train.drop(['Name'], axis=1)
df_test  = df_test.drop(['Name'], axis=1)

# Check how many titles there are now
df_train['Title'].value_counts()
```

```python
df_test['Title'].value_counts()
```

#### 2.2 Creating `FamilySize` feature from SibSp and Parch variables

SibSp refers to the number of spouses/siblings on board and Parch refers to number of parents/children on board. These two variables both refer to the presence of family members on board the ship and so it makes sense to join them and create a new feature that counts the size of a family. To create `FamilySize` I will add `SibSp` and `Parch` and add + 1 which refers to the passenger himself.

```python
# Creating 'FamilySize' feature
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch'] + 1
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch'] + 1

# Checking how many different values this feature has
df_train['FamilySize'].value_counts()
```

```python
df_test['FamilySize'].value_counts()
```

There are several family sizes with small frequency and so I will join the family sizes into fewer groups. I will do the same thing as in the previous sub-chapter but now I will create 4 groups. `Single` will refer to passengers whose family size is equal to 1; `Small` will refer to passengers whose family size is equal to 2 or 3; `Medium` will refer to passengers whose family size is equal to 4 or 5; and `Large` will refer to passengers whose family size is bigger than 5.

I will remove the `SibSp` and `Parch` variables since they will be replaced with `FamilySize`.

```python
# Replace 1 with 'Single' 
df_train['FamilySize'].replace({1: 'Single'}, inplace=True)
df_test['FamilySize'].replace({1: 'Single'}, inplace=True)

# Replace 2 and 3 with 'Small'
df_train['FamilySize'].replace(to_replace=[2, 3], value='Small', inplace=True)
df_test['FamilySize'].replace(to_replace=[2, 3], value='Small', inplace=True)

# Replace 4 and 5 with 'Medium'
df_train['FamilySize'].replace(to_replace=[4, 5], value='Medium', inplace=True)
df_test['FamilySize'].replace(to_replace=[4, 5], value='Medium', inplace=True)

# Replace 6, 7, 8 and 11 with 'Large'
df_train['FamilySize'].replace(to_replace=[6, 7, 8, 11], value='Large', inplace=True)
df_test['FamilySize'].replace(to_replace=[6, 7, 8, 11], value='Large', inplace=True)

# Remove 'SibSp' and 'Parch' variables
df_train.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df_test.drop(['SibSp', 'Parch'], axis=1, inplace=True)

# Check if data was properly grouped
df_train['FamilySize'].value_counts()
```

```python
df_test['FamilySize'].value_counts()
```

### 4. Modelling

After having done Exploratory Data Analysis (EDA), feature engineering and feature transformation, I am ready to start implementing the machine learning algorithms.

```python
# Split data into X and y
X = df_train.drop('Survived', axis=1)
y = df_train['Survived']
```

```python
# Create dictionary with all algorithms
models = {'Logistic Regression': LogisticRegression(max_iter=1000),
          'Random Forest': RandomForestClassifier(),
          'K-Nearest Neighbors': KNeighborsClassifier(),
          'Naive Bayes': GaussianNB(),
          'Linear SVM': SVC(),
          'Decision Tree': DecisionTreeClassifier()}
```

After having initialized each of the learning algorithms, I will train them and compute the F1 score to see which algorithm is more accurate.

```python
def evaluation_predictions(models, X, y):
    # Set random seed
    np.random.seed(42)
    
    # Initialize variables
    results = []
    model_names = []
    
    for model_name, model in models.items():
        # Evaluation metrics
        accuracy = np.mean(cross_val_score(model, X, y, cv=5))
        precision = np.mean(cross_val_score(model, X, y, cv=5, scoring='precision'))
        recall = np.mean(cross_val_score(model, X, y, cv=5, scoring='recall'))
        f1_score = np.mean(cross_val_score(model, X, y, cv=5, scoring='f1'))
        
        # Append evaluation metrics to 'results'
        results.append([accuracy, precision, recall, f1_score])
        
        # Append model name to 'model_names'
        model_names.append(model_name)

    # Create DataFrame with evaluation scores
    scores = pd.DataFrame(results, 
                          index=model_names,
                          columns=['Accuracy', 'Precision', 'Recall', 'F1 score'])
    
    return scores
```

```python
evaluation_predictions(models, X, y)
```

Based on these evaluation metrics, the models that perform the best on this dataset are Logistic Regression and Random Forest. I will discard all the other models and focus on these two.

#### 4.1 Hyperparameter tuning using RandomizedSearchCV

```python
# Hyperparameter grid for Random Forest
rf_grid = {'n_estimators': np.arange(10, 100, 10),
           'max_depth': [None, 3, 5, 10],
           'min_samples_split': np.arange(2, 20, 2),
           'min_samples_leaf': np.arange(1, 20, 2),
           'max_features': [0.5, 1, 'sqrt', 'auto']}

# Hyperparameter grid for Logistic Regression
lr_grid = {'penalty': ['l1', 'l2'],
           'C': np.logspace(-4, 4, 20),
           'solver': ['liblinear']}

# Instantiate RandomizedSearchCV Random Forest model
rs_rf = RandomizedSearchCV(RandomForestClassifier(n_jobs=-1),
                           param_distributions=rf_grid,
                           n_iter=100,
                           cv=5,
                           verbose=True)

# Instantiate RandomizedSearchCV Logistic Regression model
rs_lr = RandomizedSearchCV(LogisticRegression(),
                           param_distributions=lr_grid,
                           n_iter=50,
                           cv=5,
                           verbose=True)

# Fit Random Forest model to the data
rs_rf.fit(X, y)
```

```python
#Get best hyperparameters found
rs_rf.best_params_
```

```python
# Fit Logistic Regression model to the data
rs_lr.fit(X, y)
```

```python
#Get best hyperparameters found
rs_lr.best_params_
```

```python
# Instantiate Random Forest model with best hyperparameters
best_rf = RandomForestClassifier(n_estimators=70,
                                 min_samples_split=8,
                                 min_samples_leaf=7,
                                 max_features=0.5,
                                 max_depth=10)

# Instantiate Random Forest model with best hyperparameters
best_lr = LogisticRegression(solver='liblinear',
                             penalty='l2',
                             C=1.623776739188721)

best_models = {'Logistic Regression': best_lr,
               'Random Forest': best_rf}

# Evaluate model
evaluation_predictions(best_models, X, y)
```

The Logistic Regression model had a very small improvement over the baseline model while the Random Forest model had a reasonable improvement. Both models have similar results so now I will plot Learning Curves to see the bias-variance tradeoff.

#### 4.2 Bias-variance tradeoff

```python
def learning_curves(model, model_name):
    train_sizes, train_scores, validation_scores = \
        learning_curve(model, X, y, cv=5, n_jobs=-1, random_state=42)

    train_scores_mean = np.mean(train_scores, axis=1)
    validation_scores_mean = np.mean(validation_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    validation_scores_std = np.std(validation_scores, axis=1)
    
    # Plot Learning Curves
    plt.figure(figsize=(10, 5))
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, validation_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    plt.fill_between(train_sizes, validation_scores_mean - validation_scores_std,
                         validation_scores_mean + validation_scores_std, alpha=0.1,
                         color="g")
    plt.xlabel('Number of training examples', fontsize=14)
    plt.ylabel('Accuracy Score', fontsize=14)
    plt.title(f'Learning curves - {model_name}', fontsize=16, y=1.05)
    plt.legend()
    plt.grid(True);
               
# Logistic Regression learning curves
learning_curves(best_lr, 'Logistic Regression')
```

```python
# Random Forest learning curves
learning_curves(best_rf, 'Random Forest')
```

The Random Forest model seems to be overfitting the data given the constant big difference between the training and cross-validation curves.

The Logistic Regression model seems to fit the data well and generalize better since the training and cross-validation curves are close together.

I will proceed with the Logistic Regression model given the fact that it performs the best on this dataset. I will use complementary evaluation metrics on this model.

#### 4.3 Evaluation metrics

* Confusion Matrix
* ROC curve and AUC score

```python
# Set random seed
np.random.seed(42)

# Split data into training and validation set
X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size=0.2)

# Fit model to the data
best_lr.fit(X_train, y_train)

# Predictions
y_preds = best_lr.predict(X_cv)

# Confusion matrix
conf_mat = confusion_matrix(y_cv, y_preds)

# Plot confusion matrix in heat map
plt.figure(figsize=(3, 3))
sns.heatmap(conf_mat, annot=True, cbar=False, annot_kws={"size":14})
plt.xlabel('True labels', fontsize=14)
plt.ylabel('Predicted labels', fontsize=14);
```

```python
# Plot ROC curve
plt.figure(figsize=(10, 5))
plot_roc_curve(best_lr, X_cv, y_cv)
plt.plot([0, 1], [0, 1], '--', mec='r')
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('ROC Curve', fontsize=16, y=1.05);
```

### 5. Feature Importance

#### Random Forest
